#### Variables and Types

```python
# Variables and Types
length = 5  # meters (int)
width = 3  # meters (int)
height = 2.8  # meters (float)
building_type = "Residential"  # string
zip_code = "94103"  # string, even if it looks like a number
air_conditioned = True  # boolean
```

#### Lists and Slicing

A list in Python is an ordered collection of items. Each item has a position (index), and the order in which items are stored is preserved.

* **Heterogeneous Data:** Lists can store elements of different types (numbers, strings, even other lists), making them highly flexible.
* **Indexing and Access:** Each element in a list can be accessed by its index, starting from 0. This allows direct retrieval and modification of specific items.
* **Mutability:** Lists are mutable, meaning their contents can be changed after creation — items can be added, removed, or replaced.
* **Common Use:** Lists are often used to group related data together, manage sequences of items, or track collections of information that may change over time.
* **Operations:** Lists support a variety of operations, including concatenation (joining lists), slicing (extracting sublists), and iteration (processing items one by one).

```python
room_sqm = [12, 18, 25, 9, 15]  # areas of rooms in m²
print("Room areas:", room_sqm)
```

```python
# Indexing and slicing
print("First room:", room_sqm[0])
print("Last two rooms:", room_sqm[-2:])
print("Every second room:", room_sqm[::2])
```

```python
# List methods
room_sqm.append(20)  # add new room
print("After append:", room_sqm)
```

```python
# concatenate lists
more_rooms = [30, 22]
all_rooms = room_sqm + more_rooms
print("All rooms:", all_rooms)
```

```python
room_sqm.remove(9)  # remove a room
print("After remove:", room_sqm)
```

```python
print(f"Total area: {sum(room_sqm)} m²")
print(f"Largest room: {max(room_sqm)} m²")
print(f"Smallest room: {min(room_sqm)} m²")
print(f"Number of rooms: {len(room_sqm)}")
```

```python
room_sqm.sort()  # sort values
room_sqm.sort(reverse=True)  # sort descending

print("Sorted:", room_sqm)
```

#### Dictionaries

A dictionary is a collection of key–value pairs. Each key is unique, and it maps to a corresponding value.

* **Key–Value Mapping:** Instead of accessing data by position (like in lists), dictionaries allow access by key. This makes them ideal for representing structured information.

* **Data Types for Keys and Values:**
    * Keys must be immutable types (such as strings, numbers, or tuples).
    * Values can be of any type, including other collections.

* **Unordered Collection:** Dictionaries do not maintain a fixed order of items (though in modern Python, insertion order is preserved). The focus is on quick lookup by key.

* **Mutability:** Dictionaries are mutable, so new key–value pairs can be added, existing ones updated, or removed.

* **Common Use:** Dictionaries are often used when data needs to be labeled or associated, such as storing attributes, configurations, or structured records.

* **Operations:** Dictionaries support operations like retrieving values by key, checking for the existence of a key, and iterating through keys, values, or both.

```python
building_data = {
    "Building type": "Office",
    "Sqm": 1000,
    "Height": 18.2,
    "Air conditioned": True,
}
print("Dictionary:", building_data)
```

```python
# Accessing values
print("Building size:", building_data["Sqm"])
print("Building height:", building_data["Height"])
print("Is air conditioned:", building_data["Air conditioned"])
```

```python
# Adding new key-value pair
building_data["Cooling Set Point"] = 25
print("Updated dict:", building_data)
```

#### Loops and Logical Statements

**Logical Statements** <br>
Logical statements are expressions that evaluate to `True` or `False`. They are used to control the flow of a program.

* Comparison operators (e.g., ==, !=, <, >) test relationships between values.
* Logical operators (and, or, not) combine or modify conditions.

**Conditional Execution** <br>
With if, elif, and else statements, Python can make decisions. The program chooses a path of execution depending on whether conditions evaluate as True or False.

**Loops** <br>
Loops allow repetition of tasks without rewriting code:
* For loops iterate over sequences (like lists, strings, or ranges), executing a block of code for each element.
* While loops repeat a block of code as long as a condition remains True.

**Common Use** <br>
Loops and logical statements are combined to automate decision-making and repetition, forming the foundation of program control flow.

```python
# Loop through list
for area in room_sqm:
    print(area, "m²")
```

```python
# Loop through keys and values
for key, value in building_data.items():
    print(key, "->", value)
```

```python
# if-elif-else statements
if building_data["Sqm"] > 500:
    print("Large building")
elif building_data["Sqm"] > 100:
    print("Medium building")
else:
    print("Small building")
```

```python
# Combining loops and conditions
for area in room_sqm:
    if area < 10:
        print(area, "m² -> Too small!")
    elif area < 20:
        print(area, "m² -> Medium")
    else:
        print(area, "m² -> Large")
```

#### Functions and Modules

**Functions** <br>
A function is a reusable block of code that performs a specific task. Instead of repeating instructions, you define them once and call the function whenever needed.

* Parameters and Arguments: Functions can take inputs (parameters) that customize their behavior. When calling the function, the actual values passed in are called arguments.

* Return Values: Functions can send back (return) a result to the part of the program that called them, enabling reuse of computed results.

* Abstraction and Reusability: Functions make programs clearer and more maintainable by hiding details and grouping related instructions under a single name.

**Modules** <br>
A module is a file containing Python definitions (functions, classes, variables) that can be imported into other programs. Modules help organize code into manageable, reusable components.

* The Standard Library: Python includes a wide range of built-in modules (e.g., `math`, `datetime`, `random`) that provide ready-made functionality, reducing the need to reinvent common solutions.

* Importing Modules: With the import statement, you can access external code from Python’s standard library or third-party packages, extending your program’s capabilities.


<br>
Functions handle repeated or complex tasks, while modules organize and share code across projects. Together, they form the foundation for building structured, scalable programs.

```python
def energy_use_intensity(energy_use, floor_area):
    """Calculate Energy Use Intensity (EUI) in kWh/m²
    energy_use: in kWh
    floor_area: in m²
    """
    eui = energy_use / floor_area
    return eui
```

```python
eui_value = energy_use_intensity(12000, 200)  # 12000 kWh, 200 m²
print("EUI:", eui_value)
```

```python
# Importing a core Python module
import math

circle_area = math.pi * (5**2)  # Area of circle with radius
print("Circle area with radius 5:", circle_area)
```

## Part 2: Pandas Basics / Working with DataFrames

**Pandas Library** <br>
Pandas is a Python library designed for working with structured data. It provides high-level tools to clean, organize, and analyze data efficiently.
* Series: A Series is a one-dimensional labeled array, similar to a column in a table. Each value has an associated label (index).
* DataFrame: A **DataFrame** is a two-dimensional table made up of rows and columns. Each column is a Series, and together they form a structured dataset.
* Indexing and Labels: Unlike lists or arrays that rely on numeric positions, DataFrames use labels (row and column names) for intuitive data access and alignment.

**Data Handling** <br>
DataFrames make it easy to:
* Select and filter subsets of data.
* Handle missing values.
* Add, remove, or transform columns.
* Perform group-based operations and aggregations.

**Integration** <br>
Pandas integrates with many file formats and data sources (CSV, Excel, SQL, JSON, etc.), making it a powerful tool for real-world data workflows.

**Vectorized Operations** <br>
Operations on DataFrames are optimized to apply across entire columns or rows at once, making computations more efficient than manual loops.

**Common Use** <br>
DataFrames are widely used for data cleaning, exploration, and preparation before visualization or deeper analysis.

```python
import pandas as pd
```

```python
# Create a dataset of monthly energy usage (kWh)
data = {
    "Month": [
        "Jan",
        "Feb",
        "Mar",
        "Apr",
        "May",
        "Jun",
        "Jul",
        "Aug",
        "Sep",
        "Oct",
        "Nov",
        "Dec",
    ],
    "Energy_kWh": [320, 280, 300, 250, 200, 180, 170, 190, 220, 260, 300, 330],
    "Temperature_C": [2, 4, 8, 12, 18, 22, 25, 24, 20, 14, 8, 4],
}

df = pd.DataFrame(data)
print(df.head())
```

```python
# Summary statistics
print(df.describe())
```

```python
# Column selection
print(df["Energy_kWh"].head())
```

```python
# Row selection (slicing)
print(df.iloc[0:3])  # first three rows
```

```python
# Slicing backwards
print(df.iloc[-3:])  # last three rows
```

```python
# Conditional filtering
summer = df[df["Temperature_C"] > 20]
print("Summer months:")
print(summer)
```

```python
# Filtering with multiple conditions
moderate = df[(df["Temperature_C"] > 10) & (df["Energy_kWh"] < 250)]
print("Moderate weather, low energy use:")
print(moderate)
```

```python
# Add new column
df["Energy_per_degree"] = df["Energy_kWh"] / df["Temperature_C"]
print(df.head())
```

### Data Preprocessing

```python
# Drop a column
df = df.drop(columns=["Energy_per_degree"]) # reassign to df to keep the change
print(df.head())
```

```python
# Drop rows with NaN values
example_df = pd.DataFrame({"A": [1, 2, None, 4], "B": ["x", None, "y", "z"]})
print("Original:")
print(example_df)

clean_df = example_df.dropna()
print("After dropna:")
print(clean_df)
```

```python
# Fill missing values
filled_df = example_df.fillna({"A": 0, "B": "missing"})
print("After fillna:")
print(filled_df)
```

```python
# Merge DataFrames
materials = pd.DataFrame(
    {"Material": ["Brick", "Concrete", "Glass"], "ThermalConductivity": [0.6, 1.7, 0.8]}
)

prices = pd.DataFrame(
    {"Material": ["Brick", "Concrete", "Glass"], "Price_per_unit": [50, 80, 120]}
)

merged = pd.merge(materials, prices, on="Material")
print("Merged DataFrame:")
print(merged)
```

```python
# Loop with DataFrames
for idx, row in df.iterrows():
    if row["Energy_kWh"] > 300:
        print(row["Month"], "-> High energy use")
```

```python
# Using functions with DataFrames
def classify_energy_use(energy):
    if energy > 300:
        return "High"
    elif energy > 200:
        return "Medium"
    else:
        return "Low"
    
df["Energy_Class"] = df["Energy_kWh"].apply(classify_energy_use)

df
```

```python
# Exporting DataFrames
df.to_csv("data/export/monthly_energy_usage.csv", index=False)
```

## Part III: Data Visualization

**Purpose of Visualization** <br>
Data visualization transforms raw data into graphical representations. This makes patterns, trends, and relationships easier to understand than through numbers alone.

**Matplotlib Library** <br>
Matplotlib is a foundational Python library for creating static, animated, and interactive visualizations. It provides fine-grained control over all elements of a plot.

Basic Components of a Plot <br>
* Figure → the overall container for a visualization.
* Axes → the specific area where data is drawn (with x- and y-axes).
* Data Representation → lines, bars, points, or other graphical marks.
* Labels, Titles, Legends → textual elements that provide context.

**Types of Plots** <br>
Matplotlib supports many standard visualization types, such as line plots, bar charts, scatter plots, and histograms, each suited for different kinds of data analysis.

**Customization** <br>
Almost every visual element (colors, markers, scales, fonts, etc.) can be adjusted to highlight specific aspects of the data.

**Integration with pandas** <br>
Pandas works seamlessly with matplotlib, allowing quick plotting directly from DataFrames and Series, making exploratory data analysis faster.

**Common Use** <br>
Matplotlib is widely used for producing publication-quality figures, prototyping data visualizations, and serving as the basis for higher-level libraries (like seaborn or plotly).

```python
import matplotlib.pyplot as plt
```

```python
# Line plot
plt.figure(figsize=(8, 5))
plt.plot(df["Month"], df["Energy_kWh"], marker="o")
plt.title("Monthly Energy Consumption")
plt.xlabel("Month")
plt.ylabel("Energy (kWh)")
plt.show()
```

```python
# Bar chart
plt.figure(figsize=(8, 5))
plt.bar(df["Month"], df["Energy_kWh"], color="skyblue")
plt.title("Energy Consumption by Month")
plt.xlabel("Month")
plt.ylabel("Energy (kWh)")
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 5))
plt.hist(df["Energy_kWh"], bins=8, edgecolor="black")
plt.title("Distribution of Energy Consumption")
plt.xlabel("Energy (kWh)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df["Temperature_C"], df["Energy_kWh"])
plt.title("Energy Use vs Temperature")
plt.xlabel("Temperature (°C)")
plt.ylabel("Energy (kWh)")
plt.show()
```

```python
# Adding a trend line to scatter plot
import numpy as np

m, b = np.polyfit(df["Temperature_C"], df["Energy_kWh"], 1)
plt.figure(figsize=(8, 5))
plt.scatter(df["Temperature_C"], df["Energy_kWh"])
plt.plot(df["Temperature_C"], m * df["Temperature_C"] + b, color="red")
plt.title("Energy Use vs Temperature with Trend Line")
plt.xlabel("Temperature (°C)")
plt.ylabel("Energy (kWh)")
plt.show()
```
